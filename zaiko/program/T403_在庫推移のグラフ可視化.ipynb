```python
def count_jundate(data, part_number, start_time, end_time):
    
    # 検収日時をdatetimeオブジェクトに変換
    data['更新日時'] = pd.to_datetime(data['更新日時'])
    # 品番をstrオブジェクトに変換
    data['品番'] = data['品番'].astype(str)
    data['品番'] = [string.replace(" ", "") for string in data['品番']]
    #print(data['品番'])
    
    #print(part_number,type(part_number),type(data['品番']),type(data['検収日時']))
    #part_number = str(part_number)
    #print(part_number,type(part_number))
    
    #print( part_number == data['品番'])
    
    # 指定した品番と時刻のデータを抽出
    filtered_data = data[(data['品番'] == part_number) & (data['更新日時'] >= start_time) & (data['更新日時'] <= end_time)]
    #filtered_data = data[(data['収容数'] == 2000) & (data['検収日時'] >= start_time) & (data['検収日時'] <= end_time)]
    #filtered_data = data[(data['品番'] == part_number)]
    #filtered_data = data[(data['検収日時'] >= start_time) & (data['検収日時'] <= end_time)]
    
    #print(filtered_data)

    # 1時間単位でカウント
    counted_data = filtered_data
    #counted_data = filtered_data.groupby(pd.Grouper(key='更新日時', freq='1H')).size()
    #print(counted_data)
    return counted_data


def search_ABC(data, part_number):
    flag = 0
    data['品番'] = data['品番'].astype(str)
    #print(data['品番'])
    data['品番'] = [string.replace("-", "") for string in data['品番']]
    #print(data['加工図符号'])
    filtered_data = data[(data['品番'] == part_number)]
    #print(filtered_data)
    #print(part_number)
    if part_number in data['品番'].values:
        A=filtered_data['納入ｻｲｸﾙ間隔']
        B=filtered_data['納入ｻｲｸﾙ回数']
        C=filtered_data['納入ｻｲｸﾙ情報']
        #print(A,B,C)
        #if (not A) or ( not B) or (not C):
        flag = 1
        return float(A),float(B),float(C),flag
    else:
        #print(part_number,"無し")
        return 0,0,0,flag

def search_nichiryouuu(data,part_number):
    data['品番'] = data['品番'].astype(str)
    data['品番'] = [string.replace("-", "") for string in data['品番']]
    filtered_data = data[(data['品番'] == part_number)]
    x = filtered_data['当月必要数'].str.replace(',', '').astype(float)
    y = filtered_data['当月稼働日数'].astype(float)
    return float(x),float(y)


def search_hako(data,part_number):
    data['品番'] = data['品番'].astype(str)
    
    filtered_data = data[(data['品番'] == part_number)]
    filtered_data2= filtered_data['収容数'].str.replace(',', '').astype(float)
    x=float(filtered_data2)
    return x
    
def judge_kata(data,MAX):
    flag = 0;
    for i in data.values:
        if i > MAX:
            flag = 1
    return flag

def judge_keppinn(data,MIN):
    flag = 0;
    for i in data.values:
        if i < MIN:
            flag = 1
    return flag

#二つかんばんサイクルがあるものは削除
data2 = pd.read_csv('..//data//XR10_手配運用情報230712 1Y_data2.csv',encoding='cp932')
outputdata = pd.read_csv('..//data//集計結果_過多欠品.csv',encoding='shift_jis')

number = data.iloc[:,1].unique()

# 品番と時刻の範囲を指定
for i in range(len(number)):
    part_number = number[i]#data.iloc[i,2]#'356680LC020    '
    start_time  = pd.to_datetime('2023-01-01 00:00:00')
    end_time    = pd.to_datetime('2023-06-21 00:00:00')
    fig=plt.figure(figsize=[10,5])
    #print(part_number)

    #カウント
    counted_data = count_jundate(data, part_number, start_time, end_time)
    #print(counted_data)

    #
    A,B,C,flag = search_ABC(data2,part_number)
    #print(A,B,C)
    
    if flag == 1:
        ax=fig.add_subplot(111)
        
        x4,nx4 = search_nichiryouuu(data_mouth_4,part_number)
        x5,nx5 = search_nichiryouuu(data_mouth_5,part_number)
        nr = (x4/nx4+x5/nx5)/2
        
        hako = search_hako(data_mouth_4,part_number)

        # 可視化
        plt.plot(counted_data['更新日時'],counted_data['現在在庫（箱）'],label="順立装置の在庫数")
        #plt.plot(counted_data4,label="組立回収数")
        plt.xticks(rotation=90)
        plt.xlabel('Time')
        plt.ylabel('Count')
        #日糧数に変更！！！
        #MIN=0.1*(counted_data['現在在庫（箱）'].mean()*A*(1+C)/B)
        MIN=0.1*((nr//hako+1)*A*(1+C)/B)
        MAX=MIN+(nr//hako+1)/B
        #print(part_number,"日量数",nr,"収容数",hako,"A",A,"B",B,"C",:C
        ax.axhline(MIN,ls="--",color="magenta")
        ax.axhline(MAX,ls="--",color="green")
        #print(counted_data['現在在庫（箱）'])
        print("MIN:",MIN,"MAX:",MAX,"日量数:",nr,"A:",A,"B:",B,"C:",C,"日量数*A*(1+C)/B:",nr*A*(1+C)/B)
        plt.ylim( 0, MAX*2)
        plt.legend(loc="upper right",prop={"family":"MS Gothic"})
        plt.title('Count Trend for Part Number: ' + part_number + ' (From ' + str(start_time) + ' to ' + str(end_time) + ')')

        if ('*' in part_number)==False: 
            filename = 'result2/{}.png'.format(part_number)
        else:
            filename = 'result2/{}.png'.format(i)
        plt.savefig(filename)
        
        outputdata.loc[i,'1Y']=part_number
        flag_kata = judge_kata(counted_data['現在在庫（箱）'],MAX)
        flag_keppinn = judge_keppinn(counted_data['現在在庫（箱）'],MIN)
        
        outputdata.loc[i,'過多']=flag_kata
        outputdata.loc[i,'欠品']=flag_keppinn
        
        plt.show()
    else:
        outputdata.loc[i,'1Z']=part_number
        
    #outputdata.head(10)
    #outputdata.to_csv('..//data//集計結果_過多欠品.csv',encoding='shift_jis')
```
